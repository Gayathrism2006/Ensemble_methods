```python
from sklearn.datasets import load_breast_cancer
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
import pandas as pd
data=load_breast_cancer()
x_data=data.data
y_target=data.target
x=pd.DataFrame(x_data,columns=data.feature_names)
y=pd.DataFrame(y_target,columns=['target'])
print(x.head())
print(y.head())
print(x.shape)
print(y.shape)
```

```python
from sklearn.ensemble import StackingClassifier
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
scaler=StandardScaler()
x_strain=scaler.fit_transform(x_train)
x_stest=scaler.transform(x_test)
base_model=[('dc',DecisionTreeClassifier()),('lr',LogisticRegression()),('svc',SVC())]
meta_model=LogisticRegression()
model=StackingClassifier(estimators=base_model,final_estimator=meta_model)
model.fit(x_strain,y_train.values.ravel())
predict=model.predict(x_stest)
test=model.score(x_stest,y_test)
train=model.score(x_strain,y_train)
print(" Simple Stacking accuracy:",accuracy_score(y_test.values.ravel(),predict))
print("Testing accuracy:",test)
print("Training accuracy:",train)
```

```python
from sklearn.ensemble import StackingClassifier
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
scaler=StandardScaler()
x_strain=scaler.fit_transform(x_train)
x_stest=scaler.transform(x_test)
base_model=[('dc',DecisionTreeClassifier()),('lr',LogisticRegression()),('svc',SVC())]
meta_model=LogisticRegression()
model=StackingClassifier(estimators=base_model,final_estimator=meta_model,cv=8)
model.fit(x_strain,y_train.values.ravel())
predict=model.predict(x_stest)
test=model.score(x_stest,y_test)
train=model.score(x_strain,y_train)
print(" Simple Stacking accuracy:",accuracy_score(y_test.values.ravel(),predict))
print("Testing accuracy:",test)
print("Training accuracy:",train)
```
