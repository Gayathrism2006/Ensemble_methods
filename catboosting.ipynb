```python
from catboost import CatBoostClassifier
import pandas as pd
import numpy as np
from catboost.datasets import titanic
df,_ = titanic()
x = df.drop("Survived", axis=1)
y = df["Survived"]
x_data=pd.DataFrame(x,columns=x.columns)
y_data=pd.DataFrame(y,columns=['Survived'])
print(x_data.head())
print(y_data.head())

```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, random_state=42, test_size=0.2)
categorical_features_indices = [

    'Name',
    'Sex',
    'Ticket',
    'Cabin',
    'Embarked'
]
for col in categorical_features_indices:
    x_train[col] = x_train[col].astype(str)
    x_test[col] = x_test[col].astype(str)

model=CatBoostClassifier(iterations=100,learning_rate=0.1,max_depth=2, cat_features=categorical_features_indices)
model.fit(x_train,y_train, cat_features=categorical_features_indices,verbose=False)
pre=model.predict(x_test)
print("Catboost accuracy:",accuracy_score(y_test,pre))
```
